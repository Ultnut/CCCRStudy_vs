# 재규 함수
- 재귀
    - 스스로를 반복 호출 하는것
    - 나루토(꿈속에 꿈을 넣는다.)

```
n! = n * (n-1) * (n-2) * ... *1
```
4! = 4* 3* 2* 1 = 24
- 반복문을 이용해서 펙토리얼 함수를 만들어 보세요.

```
> factorial(4)
24
```

````python
result = 1
for i in range(4,0,-1):
    result = result * i
print(result)
````

````python
def factorial_1(n):
    result = 1
    for i in range(n,0,-1):
        result = result * i
    return result
````

````python
factorial_1(5)
````

````python
def factorial_2(n):
    result = 1
    for i in range(1,n+1):
        result = result * i
    return result
````

````python
factorial_2(5)
````

````python
def factorial_3(n):
    if n == 0:
        return 1
    else:
        result = n * factorial_3(n-1)
        return result
````

````python
factorial_3(3)
````

# 1부터 n 까지의 합

````python
def sum_1(n):
    if n == 0:
        return 0
    else:
        result = n + sum_1(n-1)
        return result
````

````python
sum_1(10)
````

# x^N 을 구하는 함수

````python
def power(x,n):
    if n == 0:
        return 1
    else:
        return x * power(x,n-1)
````

````python
power(3,3)
````

# 재귀함수의 문제점 
- 같은 연산을 계속 반복해서 해야하는 문제가 발생한다.
- 해결: 같은값(이미연산한 값) 한번만 계산하도록 코드를 수정 -> 메모화

````python
counter = 0
dictionary = {1:1, 2:1}
def fibonacci(n):
    print("피보나치 ({})을 계산합니다.".format(n))
    global counter
    counter += 1
    if n in dictionary:
        return dictionary[n]
    else:
        result = fibonacci(n-1) + fibonacci(n-2)
        dictionary[n] = result
        return result
fibonacci(10)   
print(counter)
````
